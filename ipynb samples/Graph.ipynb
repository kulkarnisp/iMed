```python
import wfdb
import scipy.signal as signal
import peakutils
import numpy as np
import os
import plotly.graph_objects as go
from scipy.signal import find_peaks
from matplotlib import pyplot as plt


class loader1:
    def loader_(self,rpath):
        self.records={}
        self.p={}
        self.q={}
        self.r={}
        self.s={}
        self.t={}
        self.signal_slice={}
        self.smoothed_signal={}
        self.r_peaks={}
        flist = os.listdir(rpath)
        flist = [file for file in flist if file.endswith(".dat")]   
        for file in flist:
            self.records[file] = wfdb.rdrecord(rpath+file[:-4],channels=[0])
    def signal_slice_(self):
        self.signal_slice['sel100.dat']=np.ndarray.flatten(self.records['sel100.dat'].p_signal[70:40000]) #Till 70 noise is present
        self.signal_slice['sel103.dat']=np.ndarray.flatten(self.records['sel103.dat'].p_signal[80:40000]) #Till 80 noise is present
        self.signal_slice['sel16265.dat']=np.ndarray.flatten(self.records['sel16265.dat'].p_signal[0:40000])
        self.signal_slice['sel16273.dat']=np.ndarray.flatten(self.records['sel16273.dat'].p_signal[37:40000]) #From 0:37 contains T region alone.
        self.signal_slice['sel16786.dat']=np.ndarray.flatten(self.records['sel16786.dat'].p_signal[0:40000])
        self.signal_slice['sel16420.dat']=np.ndarray.flatten(self.records['sel16420.dat'].p_signal[0:40000])
        self.signal_slice['sel16483.dat']=np.ndarray.flatten(self.records['sel16483.dat'].p_signal[100:40000]) #Till 100 noise is present
        self.signal_slice['sele0104.dat']=np.ndarray.flatten(self.records['sele0104.dat'].p_signal[0:40000])

    def r_peaks_(self):
        self.r_peaks['sel100.dat']=peakutils.indexes(self.signal_slice['sel100.dat'],thres=0.5,min_dist=0.1)
        self.r_peaks['sel103.dat']=peakutils.indexes(self.signal_slice['sel103.dat'],thres=0.5,min_dist=0.1)
        self.r_peaks['sel16265.dat']=peakutils.indexes(self.signal_slice['sel16265.dat'],thres=0.5,min_dist=0.1)
        self.r_peaks['sel16273.dat']=peakutils.indexes(self.signal_slice['sel16273.dat'],thres=0.5,min_dist=0.1)
        self.r_peaks['sel16786.dat']=peakutils.indexes(self.signal_slice['sel16786.dat'],thres=0.5,min_dist=0.1)
        self.r_peaks['sel16420.dat']=peakutils.indexes(self.signal_slice['sel16420.dat'],thres=0.5,min_dist=0.1)
        self.r_peaks['sel16483.dat']=peakutils.indexes(self.signal_slice['sel16483.dat'],thres=0.5,min_dist=0.1)
        self.r_peaks['sele0104.dat']=peakutils.indexes(self.signal_slice['sele0104.dat'],thres=0.5,min_dist=0.1)
    
    def segment_sig_(self):
        
        #Abnormal Records: sel100.dat, sel103.dat - Condition: Arrhythmia
        
        #Normal Records: sel16265.dat, sel16273.dat, sel16786.dat, sel16420.dat, sel16483.dat
        
        self.segment_sig={}
        self.segment_sig['sel100.dat']={}
        self.segment_sig['sel103.dat']={}
        self.segment_sig['sel16265.dat']={}
        self.segment_sig['sel16273.dat']={}
        self.segment_sig['sel16786.dat']={}
        self.segment_sig['sel16420.dat']={}
        self.segment_sig['sel16483.dat']={}
        
        self.time={}
        self.time['sel100.dat']=[]
        self.time['sel103.dat']=[]
        self.time['sel16265.dat']=[]
        self.time['sel16273.dat']=[]
        self.time['sel16786.dat']=[]
        self.time['sel16420.dat']=[]
        self.time['sel16483.dat']=[]
        
        # Segmenting signal into individual signals and measuring the points P,Q,R,S,T against the distance at which points P,Q,R,S,T of first individual signal are present.
        
        # Normalisation has to be done.
        # For normalisation I considered "Dictionary called time".
        
        # I considered for ex: signal_slice[0:500],which had 2 cycles. I measured both cycles P,Q,R,S,T points and saw the difference between them. The difference gives the completion of 1 cycle. Thus segmented signals based on that difference.
        
    
        # I found the total r_peaks in a record. Assumed like, number of r_peaks=number of signals.
        # So for each cycle I have assigned 1 unit of time.
        # So signal varies between 0 to 1 along the x-axis.
        
        for record in range(len(self.r_peaks['sel100.dat'])-8):
            self.segment_sig['sel100.dat'][str(record)]=self.signal_slice['sel100.dat'][(200*record):(200*(record+1))]
            self.time['sel100.dat'].append(record/198)
        for record in range(len(self.r_peaks['sel103.dat'])-8):
            self.segment_sig['sel103.dat'][str(record)]=self.signal_slice['sel103.dat'][(216*record):(216*(record+1))]
            self.time['sel103.dat'].append(record/181)
        for record in range(len(self.r_peaks['sel16265.dat'])-32):
            self.segment_sig['sel16265.dat'][str(record)]=self.signal_slice['sel16265.dat'][(252*record):(252*(record+1))]
            self.time['sel16265.dat'].append(record/157)
        for record in range(len(self.r_peaks['sel16273.dat'])):
            self.segment_sig['sel16273.dat'][str(record)]=self.signal_slice['sel16273.dat'][(147*record):(147*(record+1))]
            self.time['sel16273.dat'].append(record/208)
        for record in range(len(self.r_peaks['sel16786.dat'])-30):
            self.segment_sig['sel16786.dat'][str(record)]=self.signal_slice['sel16786.dat'][(288*record):(288*(record+1))]
            self.time['sel16786.dat'].append(record/138)
        for record in range(len(self.r_peaks['sel16420.dat'])-2):
            self.segment_sig['sel16420.dat'][str(record)]=self.signal_slice['sel16420.dat'][(210*record):(210*(record+1))]
            self.time['sel16420.dat'].append(record/189)
        for record in range(len(self.r_peaks['sel16483.dat'])-206):
            self.segment_sig['sel16483.dat'][str(record)]=self.signal_slice['sel16483.dat'][(207*record):(207*(record+1))]
            self.time['sel16483.dat'].append(record/191)
            
    def peak_detect_(self):
        
        self.p['sel100.dat']=[]
        self.q['sel100.dat']=[]
        self.r['sel100.dat']=[]
        self.s['sel100.dat']=[]
        self.t['sel100.dat']=[]
        
        self.p['sel103.dat']=[]
        self.q['sel103.dat']=[]
        self.r['sel103.dat']=[]
        self.s['sel103.dat']=[]
        self.t['sel103.dat']=[]
        
        self.p['sel16265.dat']=[]
        self.q['sel16265.dat']=[]
        self.r['sel16265.dat']=[]
        self.s['sel16265.dat']=[]
        self.t['sel16265.dat']=[] 
        
        self.p['sel16273.dat']=[]
        self.q['sel16273.dat']=[]
        self.r['sel16273.dat']=[]
        self.s['sel16273.dat']=[]
        self.t['sel16273.dat']=[]
        
        self.p['sel16786.dat']=[]
        self.q['sel16786.dat']=[]
        self.r['sel16786.dat']=[]
        self.s['sel16786.dat']=[]
        self.t['sel16786.dat']=[]
        
        self.p['sel16420.dat']=[]
        self.q['sel16420.dat']=[]
        self.r['sel16420.dat']=[]
        self.s['sel16420.dat']=[]
        self.t['sel16420.dat']=[]
        
        self.p['sel16483.dat']=[]
        self.q['sel16483.dat']=[]
        self.r['sel16483.dat']=[]
        self.s['sel16483.dat']=[]
        self.t['sel16483.dat']=[]

        for sig in self.segment_sig['sel100.dat']:
            self.p['sel100.dat'].append((self.segment_sig['sel100.dat'][sig][39])/10)  # 39 is the Distance at which first P point is present.
            self.q['sel100.dat'].append((self.segment_sig['sel100.dat'][sig][67])/10)  # 67 is the Distance at which first Q point is present.
            self.r['sel100.dat'].append((self.segment_sig['sel100.dat'][sig][73])/10)  # 73 is the Distance at which first R point is present. 
            self.s['sel100.dat'].append((self.segment_sig['sel100.dat'][sig][78])/10)  # 78 is the Distance at which first S point is present.
            self.t['sel100.dat'].append((self.segment_sig['sel100.dat'][sig][163])/10) # 163 is the Distance at which first T point is present.
            
        for sig in self.segment_sig['sel103.dat']:
            self.p['sel103.dat'].append((self.segment_sig['sel103.dat'][sig][62])/10)  # Similarly for the below records....
            self.q['sel103.dat'].append((self.segment_sig['sel103.dat'][sig][96])/10)  
            self.r['sel103.dat'].append((self.segment_sig['sel103.dat'][sig][104])/10)  
            self.s['sel103.dat'].append((self.segment_sig['sel103.dat'][sig][109])/10)  
            self.t['sel103.dat'].append((self.segment_sig['sel103.dat'][sig][171])/10)
            
        for sig in self.segment_sig['sel16265.dat']:
            self.p['sel16265.dat'].append((self.segment_sig['sel16265.dat'][sig][110])/10)  
            self.q['sel16265.dat'].append((self.segment_sig['sel16265.dat'][sig][139])/10)  
            self.r['sel16265.dat'].append((self.segment_sig['sel16265.dat'][sig][147])/10)  
            self.s['sel16265.dat'].append((self.segment_sig['sel16265.dat'][sig][155])/10)  
            self.t['sel16265.dat'].append((self.segment_sig['sel16265.dat'][sig][218])/10)
            
        for sig in self.segment_sig['sel16273.dat']:
            self.p['sel16273.dat'].append((self.segment_sig['sel16273.dat'][sig][37])/10)  
            self.q['sel16273.dat'].append((self.segment_sig['sel16273.dat'][sig][61])/10)  
            self.r['sel16273.dat'].append((self.segment_sig['sel16273.dat'][sig][68])/10)  
            self.s['sel16273.dat'].append((self.segment_sig['sel16273.dat'][sig][77])/10)  
            self.t['sel16273.dat'].append((self.segment_sig['sel16273.dat'][sig][137])/10)
            
        for sig in self.segment_sig['sel16786.dat']:
            self.p['sel16786.dat'].append((self.segment_sig['sel16786.dat'][sig][56])/10)  
            self.q['sel16786.dat'].append((self.segment_sig['sel16786.dat'][sig][92])/10)  
            self.r['sel16786.dat'].append((self.segment_sig['sel16786.dat'][sig][98])/10)  
            self.s['sel16786.dat'].append((self.segment_sig['sel16786.dat'][sig][104])/10)  
            self.t['sel16786.dat'].append((self.segment_sig['sel16786.dat'][sig][172])/10)
            
        for sig in self.segment_sig['sel16420.dat']:
            self.p['sel16420.dat'].append((self.segment_sig['sel16420.dat'][sig][34])/10)  
            self.q['sel16420.dat'].append((self.segment_sig['sel16420.dat'][sig][70])/10)  
            self.r['sel16420.dat'].append((self.segment_sig['sel16420.dat'][sig][78])/10)  
            self.s['sel16420.dat'].append((self.segment_sig['sel16420.dat'][sig][83])/10)  
            self.t['sel16420.dat'].append((self.segment_sig['sel16420.dat'][sig][143])/10)
            
        for sig in self.segment_sig['sel16483.dat']:
            self.p['sel16483.dat'].append((self.segment_sig['sel16483.dat'][sig][47])/10)  
            self.q['sel16483.dat'].append((self.segment_sig['sel16483.dat'][sig][88])/10)  
            self.r['sel16483.dat'].append((self.segment_sig['sel16483.dat'][sig][95])/10)  
            self.s['sel16483.dat'].append((self.segment_sig['sel16483.dat'][sig][100])/10)  
            self.t['sel16483.dat'].append((self.segment_sig['sel16483.dat'][sig][160])/10)
            
            
    def plotter_(self):
        
        #Below plot is about 'P' point vs 'S' point's of different records 
        
        fig = go.Figure()
        
        #Abnormal data


        fig.add_trace(go.Scatter(x=p1.p['sel100.dat'], y=p1.s['sel100.dat'],
                            mode='markers',
                            name='P vs S: 100.dat(Abnormal)'))
        fig.add_trace(go.Scatter(x=p1.p['sel103.dat'], y=p1.s['sel103.dat'],
                            mode='markers',
                            name='P vs S: 103.dat(Abnormal)'))

        #Normal Data

        fig.add_trace(go.Scatter(x=p1.p['sel16265.dat'], y=p1.s['sel16265.dat'],
                            mode='markers',
                            name='P vs S: 16265.dat'))
        fig.add_trace(go.Scatter(x=p1.p['sel16273.dat'], y=p1.s['sel16273.dat'],
                            mode='markers',
                            name='P vs S: 16273.dat'))
        fig.add_trace(go.Scatter(x=p1.p['sel16786.dat'], y=p1.s['sel16786.dat'],
                            mode='markers',
                            name='P vs S: 16786.dat'))
        fig.add_trace(go.Scatter(x=p1.p['sel16420.dat'], y=p1.s['sel16420.dat'],
                            mode='markers',
                            name='P vs S: 16420.dat'))
        fig.add_trace(go.Scatter(x=p1.p['sel16483.dat'], y=p1.s['sel16483.dat'],
                            mode='markers',
                            name='P vs S: 16483.dat'))


        fig.update_layout(
            title="Point cloud:Normal heart rate and abnormal heart rate",
            xaxis_title="P",
            yaxis_title="S",
            font=dict(
                family="Courier New, monospace",
                size=18,
                color="Blue"
            )
        )

        fig.show()
```
