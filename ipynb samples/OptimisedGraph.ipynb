```python
import wfdb
import scipy.signal as signal
import peakutils
import numpy as np
import os
import plotly.graph_objects as go
import pandas as pd
from scipy.signal import find_peaks
from matplotlib import pyplot as plt


class loader:
    def loader_(self,rpath):
        self.records={}
        self.p={}
        self.q={}
        self.r={}
        self.s={}
        self.t={}
        self.rec1={}
        self.signal_slice={}
        self.smoothed_signal={}
        self.r_peaks={}
        flist = os.listdir(rpath)
        flist = [file for file in flist if file.endswith(".dat")]   
        for file in flist:
            self.records[file] = wfdb.rdrecord(rpath+file[:-4],channels=[0])
        self.rec1={
            'sel100.dat':[70,40000,8,200,198,39,67,73,78,163],
            'sel103.dat':[80,40000,8,216,181,62,96,104,109,171],
            'sel16265.dat':[0,40000,32,252,157,110,139,147,155,218],
            'sel16273.dat':[37,40000,0,147,208,37,61,68,77,137],
            'sel16786.dat':[0,40000,30,288,138,56,92,98,104,172],
            'sel16420.dat':[0,40000,2,210,189,34,70,78,83,143],
            'sel16483.dat':[100,40000,206,207,191,47,88,95,100,160]
        }
    def signal_slice_(self):
        for rec in self.rec1:
            self.signal_slice[rec]=np.ndarray.flatten(self.records[rec].p_signal[self.rec1[rec][0]:self.rec1[rec][1]])


        
    def r_peaks_(self):
        for rec in self.rec1:
            self.r_peaks[rec]=peakutils.indexes(self.signal_slice[rec],thres=0.5,min_dist=0.1)

    
    def segment_sig_(self):
        self.segment_sig={}
        self.time={}
        for rec in self.rec1:
            self.segment_sig[rec]={}

        for rec in self.rec1:
            self.time[rec]=[]

        for rec in self.rec1:
            for record in range(len(self.r_peaks[rec])-self.rec1[rec][2]):
                self.segment_sig[rec][record]=self.signal_slice[rec][(self.rec1[rec][3]*record):(self.rec1[rec][3]*(record+1))]
                self.time[rec].append(record/self.rec1[rec][4])
            
    def peak_detect_(self):
        for rec in self.rec1:
            self.p[rec]=[]
            self.q[rec]=[]
            self.r[rec]=[]
            self.s[rec]=[]
            self.t[rec]=[]
            
            for sig in self.segment_sig[rec]:
                self.p[rec].append((self.segment_sig[rec][sig][self.rec1[rec][5]])/10)  
                self.q[rec].append((self.segment_sig[rec][sig][self.rec1[rec][6]])/10)  
                self.r[rec].append((self.segment_sig[rec][sig][self.rec1[rec][7]])/10)  
                self.s[rec].append((self.segment_sig[rec][sig][self.rec1[rec][8]])/10)  
                self.t[rec].append((self.segment_sig[rec][sig][self.rec1[rec][9]])/10)
            
            
    def plotter_(self):
        
        #Below plot is about 'P' point vs 'S' point's of different records 
        
        fig = go.Figure()
        
        for rec in self.rec1:
            fig.add_trace(go.Scatter(x=p1.p[rec], y=p1.s[rec],
                            mode='markers',
                            name='P vs S:'+rec))


        fig.update_layout(
            title="Point cloud:Normal heart rate and abnormal heart rate",
            xaxis_title="P",
            yaxis_title="S",
            font=dict(
                family="Courier New, monospace",
                size=18,
                color="Blue"
            )
        )

        fig.show()
```
